## Analysis and Visualization
### We Rate Dogs Data
WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though are almost always greater than 10. 11/10, 12/10, 13/10, etc.

### Insights

#### 1. Total Number of tweets over time



From the visualization above, we can see that the number of tweets  in this account increased sharply on the onset until January 2016. After that, the trend witnessed a sharp nosedive from then until March 2016. The number of tweets from then on started decreasing gradually with varied small increases over time until July 2017. 
This is an alarming trend that should be looked into by the account owner since the dwindling number of tweets might depict a decrease in members' participation on the account.

#### 2. Most Popular Dog Names


From the visualization above, we can see that the most popular dog names are **Lucy** and **Charlie**, each with `11` entries. Closely followed by **Cooper** and **Oliver**, each with `10` entries. We then have **Lola**, **Tucker**, **Sadie** and **Penny** as the nest most popular dog names with `8` entries each.

#### 3. Most common Twitter Source


Without a doubt **Twitter for iPhone** is the most common source in this dataset with `1707` entries. 

The next source is **Vine - Make a Scene** which doesn't even come close with `90` entries. 

The remaining two entries, namely **TweetDeck** and **Twitter Web Client**, had measly figures of `11` entries each. 

This only goes to show that most people who post on this platform are iPhone users.  

#### 4. Correlation betweeen favorite_count and retweet_count 


The above scatter plot shows there's a very strong correlation between the favorite_count and retweet_count.
This isn't peculiar since more often than not a person will most likely retweet something they have favorited on twitter.
This correlation can be used by account owner to determine ways of increasing traffic on this page. We can use posts with historically high number of retweets to check the characteristics of the same and know what types of posts to encourage users to share and increase the traffic on this account.
